```python
import pandas as pd
import numpy as np
```

```python
# Load the Excel file.
df_all = pd.read_excel('Crypto_2022_2021_2020 (Autosaved).xlsx',sheet_name='2020_2021_2022_ALL',index_col=0)
df_all.head()
```

```python
# Lets find if there are null values in dataframe
print(df_all.shape) 
print(df_all.dropna(axis=0).shape) # number of rows that donot have null values
print(df_all.dropna(axis=1).shape) # number of columns that donot have null values
```

```python
# Let's check if there are any columns with all null values
columns = df_all.columns
print(columns)
```

```python
# Code to get individual number of null values in a column and its location
print(len(np.where(df_all.isna()['ISIN'])[0]))
np.where(df_all.isna()['Amount'])[0]
```

```python
# Let's write a function which will apply on the each coulmn 
def find_null_values(column_name):
  print('The number of null values in column {} are:'.format(column_name),len(np.where(df_all.isna()[column_name])[0]))
  print('And their integer location is:', np.where(df_all.isna()[column_name])[0])
```

```python
# Execute the funtion
null_values_and_their_integer_location_in_df_all = [find_null_values(y) for y in columns]
```
